```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe=DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})
```

```python
dframe
```

```python
dframe2=DataFrame({'key':['Q',"Y",'Z'],'dataset_2':[1,2,3]})
dframe2
```

```python
pd.merge (dframe,dframe2)
```

```python
pd.merge(dframe,dframe2,on='key')
```

```python
print(dframe,"\n")
print(dframe2)
pd.merge(dframe,dframe2,on='key',how='left')
```

```python
dframe2
```

```python
print(dframe,"\n")
print(dframe2)
pd.merge(dframe,dframe2,on='key',how='right')
```

```python
dframe2
```

```python
print(dframe,"\n")
print(dframe2)
pd.merge(dframe,dframe2,on='key',how='outer')
```

```python
#Many to Many Merge

dframe3=DataFrame({'key':['X','X','X','Y','Z','Z'],'data_set3':range(6)})

dframe3
```

```python
dframe4=DataFrame({'key':['Y',"Y",'X','X','Z'],'data_set4':np.arange(5)})

dframe4
```

```python
pd.merge(dframe3,dframe4)
```

```python
df_left=DataFrame({'key1':['SF','SF','LA'],'key2':['one','two','one'],'left_data':[10,20,30]})

df_left
```

```python
df_right=DataFrame({'key1':['SF','SF','LA','LA'],'key2':['One','One','One','Two'],'right_data':[40,50,60,70]})

df_right
```

```python
pd.merge(df_left,df_right,on=['key1','key2'],how='outer')
```

```python
pd.merge(df_left,df_right,on='key1')
```

```python
pd.merge(df_left,df_right,on='key1',suffixes=('_left','_right'))
```
